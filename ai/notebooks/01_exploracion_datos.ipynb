# 📊 Exploración del Dataset LSC

En este notebook analizaremos visualmente la calidad y distribución del dataset recolectado para el proyecto **Kairos**.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Configuración de estilo
sns.set_theme(style="whitegrid")
plt.rcParams['figure.figsize'] = (12, 6)
```

## 1. Carga y Verificación básica

```python
DATASET_PATH = '../datasets/lsc_dataset.csv'
df = pd.read_csv(DATASET_PATH)
print(f"Total de muestras: {len(df)}")
df.head()
```

## 2. Distribución de clases

Es vital que el dataset esté balanceado (el objetivo es ~50 muestras por seña).

```python
counts = df['label'].value_counts().sort_index()
sns.barplot(x=counts.index, y=counts.values, palette="viridis")
plt.xticks(rotation=45)
plt.title("Distribución de muestras por seña LSC")
plt.xlabel("Seña")
plt.ylabel("Cantidad de muestras")
plt.show()
```

## 3. Análisis de Keypoints

MediaPipe entrega los puntos normalizados entre 0 y 1. Vamos a verificar el rango de los datos.

```python
keypoints_df = df.drop('label', axis=1)
print(f"Mínimo: {keypoints_df.min().min():.4f}")
print(f"Máximo: {keypoints_df.max().max():.4f}")

# Histograma de un punto específico (ej: kp_0)
sns.histplot(df['kp_0'], kde=True)
plt.title("Distribución del keypoint 0 (X de la muñeca)")
plt.show()
```
